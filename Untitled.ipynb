```python
df['CTC']=df['CTC'].apply(lambda x: float(str(x).split()[0].replace(',', '')))
df1['Actual CTC']=df1['Actual CTC'].apply(lambda x: float(str(x).split()[0].replace(',', '')))
df['Previous CTC']=df['Previous CTC'].apply(lambda x: float(str(x).split()[0].replace(',', '')))
df1['previous CTC']=df1['previous CTC'].apply(lambda x: float(str(x).split()[0].replace(',', '')))


```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['College']=le.fit_transform(df['College'])
df['Role']=le.fit_transform(df['Role'])
df['City type']=le.fit_transform(df['City type'])
```

```python
df.head()
```

```python
df1['College']=le.fit_transform(df1['College'])
df1['Role']=le.fit_transform(df1['Role'])
df1['City type']=le.fit_transform(df1['City type'])
```

```python
x=df.drop('CTC',axis=1).values
y=df['CTC'].values
x.reshape(-1,1)
y.reshape(-1,1)
```

```python
x.shape[0]
```

```python
y.shape
```

```python
len(x)==len(y)
```

```python
from sklearn.preprocessing import RobustScaler
rs=RobustScaler()
rs.fit(x)
x1=rs.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val=train_test_split(x1,y,test_size=0.2,random_state=0)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_val)
```

```python
pred
```

```python


y_val
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
print('r2 score is :',r2_score(y_val,pred))


```

```python
print('MSE score is :',mean_squared_error(y_val,pred))
```

```python
rmse=np.sqrt(mean_squared_error(y_val,pred))
print(rmse)
```
